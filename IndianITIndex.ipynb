## Normalization
To **normalize** the stock prices divide each price by the first price and then multiplying by 100. This operation essentially scales the prices so that the first price becomes 100, and subsequent prices are relative to that.

```python
norm = stocks.div(stocks.iloc[0]).mul(100)
```

```python
norm.plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 15)
plt.title('Normalized Closing Stock Price over Time',fontsize = 15)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.grid(True)
plt.show()
```

## Creating a Price-weighted Index
**price-weighted index** is such that its constituents are **weighted** by their stock prices.  

Here are some of the pros of **Price-Weighted Index:**
1. **Simple Calculation:** Price-weighted indices are straightforward to calculate. They simply sum up the prices of all the constituent stocks and divide by a divisor (typically the number of stocks), making the calculation process easy to understand and implement.
2. **Reflect Market Movements:** Since higher-priced stocks have a greater impact on the index value, price-weighted indices can accurately reflect the movements of those stocks. This can be advantageous in markets where higher-priced stocks have a significant influence on market sentiment.
3. **Transparency:** The calculation methodology of price-weighted indices is transparent and easy to comprehend, making it easier for investors to understand how changes in stock prices impact the index value.

```python
stocks.head()
```

```python
norm.head()
```

The code below calculates the price-weighted index from the sum of the closing prices of all stocks for each date, normalizes it to have an initial value of 100, and then assigns it to a new column named "PWI" in the DataFrame norm.

```python
norm["PWI"] = stocks.sum(axis = 1).div(stocks.sum(axis = 1).iloc[0]).mul(100) 
```

```python
norm.head()
```

```python
norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.title('Normalized Closing Stock Price over Time',fontsize = 15)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.grid(True)
plt.show()
```

**Cons of Price-Weighted Index:**
1. **Skewed Representation:** Price-weighted indices can be skewed by the stock prices themselves. Stocks with higher prices will have a larger influence on the index value, regardless of their market capitalization or economic significance. This can lead to an inaccurate representation of the overall market.
2. **No Consideration of Market Capitalization:** Price-weighted indices do not take into account the market capitalization of individual stocks. As a result, they may not accurately reflect the relative importance of each stock in the market.
3. **Less Diversification:** Price-weighted indices may lack diversification because higher-priced stocks tend to have a larger impact on the index value. This can lead to an overrepresentation of certain sectors or industries in the index.
4. **Volatility:** Price-weighted indices can be more volatile compared to other types of indices, especially when high-priced stocks experience large price movements. This volatility may not accurately reflect the overall market sentiment or economic conditions.

## Creating a Equal Weighted Index
To create an equal-weighted index, you allocate an equal amount of capital to each stock in the index.   
Here are some of the **Pros** of **Equal Weighted Index**:
1. **Simplicity:** Equal-weighted indices are straightforward and easy to understand. Every stock in the index carries the same weight, which simplifies the construction and maintenance of the index.
2. **Diversification:** Equal-weighted indices inherently provide greater diversification compared to market-cap weighted indices. Since each stock has the same weight, smaller stocks have a proportionally larger impact on the index, reducing concentration risk.
3. **Potential for Outperformance:** Equal-weighted indices may outperform market-cap weighted indices during certain market conditions, especially when smaller stocks or sectors outperform larger ones. This is because equal-weighting gives more weight to smaller stocks, which have higher growth potential.  

Here are some of the **Cons** of **Equal Weighted Index**:  

1. **Higher Costs:** Rebalancing an equal-weighted index involves buying more of underperforming stocks and selling more of outperforming stocks. This can result in higher trading costs compared to market-cap weighted indices, where adjustments are made based on changes in market capitalization.
2. **Lack of Consideration for Fundamental Factors:** Equal-weighted indices do not consider fundamental factors such as company size, profitability, or liquidity. Market-cap weighted indices, on the other hand, reflect the market's valuation of companies based on their market capitalization.
3. **Volatility:** Equal-weighted indices can exhibit higher volatility compared to market-cap weighted indices. This is because smaller stocks, which tend to be more volatile, have the same weight as larger, less volatile stocks.

```python
# Calculate the daily percentage changes
ret = stocks.pct_change().dropna()
```

```python
ret.head()
```

```python
# Calculate the mean daily return for each row (each day)
ret["Mean_ret"] = ret.mean(axis = 1)
ret.head()
```

```python
norm["EWI"] = 100
norm.head()
```

```python
# We calculate the cumulative returns for the equal-weighted index based on the mean daily returns and assign them to the 'EWI' column in the 'norm' DataFrame.
norm.iloc[1:, -1] = ret.Mean_ret.add(1).cumprod().mul(100)
norm.head()
```

```python
norm.iloc[:, [0, 1, 2, 3, 4, 6]].plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.title('Normalized Closing Stock Price over Time',fontsize = 15)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.show()
```

```python
norm.loc[:, ["LTIM.NS", "EWI"]].plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.title('Normalized Closing Stock Price over Time',fontsize = 15)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.show()
```

## Creating a Value-weighted or Market Cap-Weighed Index
This type of index is a stock market index in which each component of the index is weighted according to its market capitalization, i.e., the total market value of its outstanding shares. In a market-cap weighted index, companies with higher market capitalizations will have a greater impact on the index's value compared to those with lower market capitalizations. This reflects the relative size and importance of the companies within the index.  

Get data realated to **outstanding shares** of the IT stocks in the created Index. 

```python
import csv
# Define the ticker symbols for Indian IT stocks
ticker_symbols = ["TCS.NS", "INFY.NS", "HCLTECH.NS", "LTIM.NS", "TECHM.NS"]
# Create a list to store the data
data = []

# Iterate through each ticker symbol
for ticker_symbol in ticker_symbols:
    # Download the stock data
    stock_data = yf.Ticker(ticker_symbol)
    
    # Get the outstanding shares data
    outstanding_shares = stock_data.info.get("sharesOutstanding")
    
    # Append the data to the list
    data.append({"Ticker": ticker_symbol, "OutstandingShares": outstanding_shares})

# Specify the filename
filename = "indian_it_stocks_outstanding_shares.csv"

# Write the data to a CSV file
with open(filename, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["Ticker", "OutstandingShares"])
    
    # Write the header
    writer.writeheader()
    
    # Write the data
    writer.writerows(data)
```

**Pros of Market Cap Weighed Index:**

1. **Reflects Market Dynamics:** Market cap-weighted indices reflect the market dynamics accurately by giving higher weights to larger companies. This aligns with the market's valuation of companies, as larger companies typically have a greater impact on the overall market.
2. **Low Turnover:** Since changes in the index composition only occur when there are changes in the market capitalization of the constituents, market cap-weighted indices generally have lower turnover compared to other weighting schemes. This can result in lower trading costs for index funds tracking these indices.
3. **Simple and Transparent:** Market cap-weighted indices are easy to understand and calculate. The methodology is straightforward, making it transparent for investors to know which companies are included and their respective weights in the index.
4. **Diversification:** By including a large number of stocks from various sectors, market cap-weighted indices provide a diversified exposure to the overall market, reducing individual company risk.

**Cons of Market Cap Weighed Index:**

1. **Overvaluation of Overpriced Stocks:** Market cap-weighted indices tend to overweight overvalued stocks and underweight undervalued stocks. This can lead to a situation where the index overemphasizes stocks that may be trading at high prices relative to their intrinsic value.
2. **Concentration Risk:** Since market cap-weighted indices give higher weights to larger companies, they can become heavily concentrated in a few big names or sectors. This concentration may expose investors to higher risk if those companies or sectors underperform.
3. **Limited Exposure to Small Caps:** Market cap-weighted indices may underrepresent small-cap stocks since their weights are determined by their market capitalization. This could lead to missed opportunities for investors looking to capture potential growth in smaller companies.
4. **Impact of Market Bubbles:** During market bubbles, market cap-weighted indices may become skewed towards the overvalued stocks driving the bubble, which could increase the index's vulnerability to market downturns when the bubble bursts.

```python
summary = norm.pct_change().dropna().agg(["mean", "std"]).T
```

```python
summary.head()
```

```python
summary.columns = ["Return", "Risk"]
```

```python
summary["Return"] = summary["Return"]*252
summary["Risk"] = summary["Risk"] * np.sqrt(252)
```

```python
summary.head()
```

Below is the graph shows the scatter plot of Annual Return VS Annual Risk of Individual stocks and various index such as PWI, EWI and VWI. 

```python
summary.plot(kind = "scatter", x = "Risk", y = "Return", figsize = (15, 8), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "Risk"]+0.002, summary.loc[i, "Return"]+0.002), size = 15)
plt.xlabel("Annual Risk(std)", fontsize = 15)
plt.ylabel("Annual Return", fontsize = 15)
plt.title("Risk/Return Scatter Plot", fontsize = 20)
plt.show()
```

From the above the graph we see that all the index like **PWI, EWI and VWI** are **towards lower end annual risk.** While individual stocks have higher returns and higher risks than the Index. From this we can conclude that investing in an index of stocks reduces the overall risk but it also reduces the annual returns.
